```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### 데이터 가져오기

```python
train = pd.read_csv("~/data/data/bike-sharing-demand/train.csv")
```

### datetime 컬럼을 datetime 자료형으로 변환하고 연, 월, 일, 시, 분, 초까지 6가지 컬럼 생성하기

```python
train["datetime"] = pd.to_datetime(train["datetime"])
train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
train["minute"] = train["datetime"].dt.minute
train["second"] = train["datetime"].dt.second
train.head()
```

### year, month, day, hour, minute, second 데이터 개수 시각화하기

```python
plt.figure(figsize=(12,7))
plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.35)
plt.subplot(2,3,1)
sns.countplot(data=train, x="year")
plt.title("Count plot by year")
plt.subplot(2,3,2)
sns.countplot(data=train, x="month")
plt.title("Count plot by month")
plt.subplot(2,3,3)
sns.countplot(data=train, x="day")
plt.title("Count plot by day")
plt.subplot(2,3,4)
sns.countplot(data=train, x="hour")
plt.title("Count plot by hour")
plt.subplot(2,3,5)
sns.countplot(data=train, x="minute")
plt.title("Count plot by minute")
plt.subplot(2,3,6)
sns.countplot(data=train, x="second")
plt.title("Count plot by second")
```

### X, y 컬럼 선택 및 train/test 데이터 분리

```python
X = train.drop(columns="count")
y = train["count"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
### 변수 선택
drop_col = ["datetime", "casual", "registered"]
```

```python
X_train = X_train.drop(columns=drop_col)
X_test = X_test.drop(columns=drop_col)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### LinearRegression 모델 학습

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("MSE  : ", mean_squared_error(y_test, y_pred))
print("RMSE : ", mean_squared_error(y_test, y_pred, squared=False))
```

### x축은 temp 또는 humidity로, y축은 count로 예측 결과 시각화하기

```python
plt.figure(figsize=(12,5))
plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.35)

plt.subplot(1,2,1)
plt.scatter(x=X_test["temp"], y=y_test, s=1, label="True")
plt.scatter(x=X_test["temp"], y=y_pred, s=1, label="Pred")
plt.legend()
plt.title("tmep")

plt.subplot(1,2,2)
plt.scatter(x=X_test["humidity"], y=y_test, s=1, label="True")
plt.scatter(x=X_test["humidity"], y=y_pred, s=1, label="Pred")
plt.legend()
plt.title("humidity")

plt.show()
```

```python
sns.kdeplot(y_test, fill=True, label="True")
sns.kdeplot(y_pred, fill=True, label="Pred")
plt.legend()
plt.show()
```
